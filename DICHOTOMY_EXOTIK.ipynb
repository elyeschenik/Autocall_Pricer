```python
def BS_Dividends(option_start, option_maturity, option_type, S, k, r, q, sigma):
    
    t = (option_maturity - option_start).days
    D1 = (np.log((S * np.exp(-q * t / 365)) / (k * np.exp(-r * t / 365))) + (1 / 2) * (sigma ** 2) * t / 365) / (sigma * np.sqrt(t / 365))
    D2 = D1 - sigma * np.sqrt(t / 365)

  

    if option_type == "Call":
        ND1 = stats.norm.cdf(D1)
        ND2 = stats.norm.cdf(D2)
        BS = S * np.exp(-q*t/365.0)*ND1-k * np.exp(-r * t / 365.0) * ND2
    
    else:
        
        ND1 = stats.norm.cdf(-D1)
        ND2 = stats.norm.cdf(-D2)
        BS = k * np.exp(-r * t / 365) * ND2 - S * np.exp(-q * t / 365) * ND1
    
    return BS
    
    
    
```

```python
def Dichotomy(option_start, option_maturity, option_type, S, k, price, r ,q):
    
    t = (option_maturity - option_start).days
    n_iter = 0
    sigma_min = 0
    sigma_max = 5
    
    eps = 0.0001
    val = 0
    
    premium = price
    
    while abs(val - premium)> eps and n_iter < 100:
        
        sigma_s = (sigma_min + sigma_max)/2
        val =  BS_Dividends(option_start, option_maturity, option_type, S, k, r, q, sigma_s)
      
        if val <= premium:
            sigma_min = sigma_s
            sigma_max = sigma_max
         
        
        elif val >= premium:
            sigma_min = sigma_min
            sigma_max = sigma_s
        
        
        n_iter = n_iter + 1
        
        print('n_iter is {} and we have val {} vs premium {} given sigma {}'.format(n_iter, np.round(val,4), np.round(premium,4), np.round(sigma_s,4)))
        
    return sigma_s
    
```

```python
from dateutil.relativedelta import *

swap_rates['Start_Date'] = pd.Timestamp(2021,4,7)
swap_rates['End_Date'] = 0

for i in range(len(swap_rates)):
    if i < 10:
        if swap_rates['period'][i][-1:] == 'Y':
            swap_rates['End_Date'][i] = swap_rates['Start_Date'][i] + relativedelta(years=+ int(swap_rates['period'][i][:1]))
        if swap_rates['period'][i][-1:] == 'M':
            swap_rates['End_Date'][i] = swap_rates['Start_Date'][i] + relativedelta(months=+ int(swap_rates['period'][i][:2]))
            
    if i >= 10:
         if swap_rates['period'][i][-1:] == 'Y':
            swap_rates['End_Date'][i] = swap_rates['Start_Date'][i] + relativedelta(years=+ int(swap_rates['period'][i][:2]))
        
swap_rates.set_index('End_Date', inplace = True)
```

```python
def swap_rate(T):
    if T in swap_rates.index:
        R = swap_rates.loc[T]['EUR Swap Curve']
    else:
        if T > pd.Timestamp(2022, 4, 7):
            swap_rates.loc[T] = None
            R = swap_rates.sort_index().interpolate(method='time').loc[T]['EUR Swap Curve']
            swap_rates.drop(T, axis = 0, inplace = True)
        else:
            #R = -0.6 
            R = swap_rates.loc[pd.Timestamp(2022, 4, 7)]['EUR Swap Curve']
    return R
```
